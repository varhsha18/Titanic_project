```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_columns', None)
```

```python
df=pd.read_csv("E:\\final_titanic _project\\titanic\\train.csv")
df1=pd.read_csv("E:\\final_titanic _project\\titanic\\test.csv")
```

```python
df.drop(columns=['Cabin'],inplace=True )
df1.drop(columns=['Cabin'],inplace=True )
```

```python
df.isnull().sum()
```

```python
df1.isnull().sum()
```

```python
k=df1['Fare'].mean()
df1['Fare'].fillna(k, inplace=True)
#df1.fillna({df1['Fare']: k}, inplace=True)
```

```python
kk1=df['Embarked'].mode()
df['Embarked'].fillna(kk1[0], inplace=True)
```

```python
import numpy as np
```

```python
train_age=np.random.randint(df['Age'].mean()-df['Age'].std(),df['Age'].mean()+df['Age'].std(),177)
test_age=np.random.randint(df1['Age'].mean()-df1['Age'].std(),df1['Age'].mean()+df1['Age'].std(),86)
```

```python
df['Age'][df['Age'].isnull()]=train_age
df1['Age'][df1['Age'].isnull()]=test_age
```

```python
df.drop(columns=['Ticket'],inplace=True )
df1.drop(columns=['Ticket'],inplace=True )
```

```python
df['family']=df['SibSp']+df['Parch']+1
df1['family']=df1['SibSp']+df1['Parch']+1
```

```python
def tit(hhh):
     if hhh==1:
         return 'Alone'
     elif hhh>1 and hhh<5:
       return 'Medium'    
     else :
       return 'Large'    
```

```python
df['family_size']=df['family'].apply(tit)
```

```python
df1['family_size']=df1['family'].apply(tit)
```

```python
df.drop(columns=['SibSp','Parch','family'],inplace=True )
```

```python

df1.drop(columns=['SibSp','Parch','family'],inplace=True )
```

```python
vv=df1['PassengerId'].values
```

```python
df.drop(columns=['Name','PassengerId'],inplace=True )
df1.drop(columns=['Name','PassengerId'],inplace=True )
```

```python
df=pd.get_dummies(df,["Sex","Embarked","family_size"],drop_first=True)
df1=pd.get_dummies(df1,["Sex","Embarked","family_size"],drop_first=True)
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=2)
```

```python
from sklearn.ensemble import RandomForestClassifier  
from sklearn.metrics import accuracy_score
model= RandomForestClassifier(n_estimators=500,max_features=0.75, min_samples_leaf=5)  
```

```python
model.fit(X_train, y_train)

#Making predictions on the test set
predictions = model.predict(X_test)

#Calculating accuracy
accuracy = accuracy_score(y_test, predictions)

print("Accuracy:", accuracy)
```

```python
y_final=model.predict(df1)
```
